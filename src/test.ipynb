```python
from data import _get_stock_filename
import mplfinance as mpf
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

def _rename_table_head(data: pd.DataFrame) -> pd.DataFrame:
    # code,name,time_key,open,close,high,low,pe_ratio,turnover_rate,volume,turnover,change_rate,last_close
    new_col_names = {
        'time_key': 'Date',
        'open': 'Open',
        'high': 'High',
        'low': 'Low',
        'close': 'Close',
        'volume': 'Volume'
    }

    data.rename(columns=new_col_names, inplace=True)
    data['Date'] = data['Date'].apply(lambda x: x.split(' ')[0])
    data['Date'] = pd.to_datetime(data['Date'])
    data.set_index('Date', inplace=True)
    return data

data = pd.read_csv(_get_stock_filename('HK.09866'), parse_dates=True)
data = _rename_table_head(data)

points =[('2022-08-15', 150), ('2023-12-12', 60)]
kwargs = {
    'type': 'candle',
    'volume': True,
    'show_nontrading': True,
    'style': 'charles',
    'title': 'K-Line',
    'ylabel_lower': 'Deal',
    'ylabel': 'Price',
    'returnfig': True
}

fig, ax = mpf.plot(data, **kwargs)
axs = fig.get_axes()

for date, price in points:
    if date in data.index:
        date = pd.to_datetime(date)
        print(f"date: {date}, price: {price}")
        ax[0].scatter(date, price, color='blue', marker='o', zorder=1, s=10)

plt.show()
```
