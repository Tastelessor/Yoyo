```python
import itertools
import numpy as np

param_stride = 0.1

# Define the 'change_rate' parameters with their respective ranges
change_rate_params = {
    "a1": [4.0, 5.0],
    "a2": [0.0, 0.0],  # This range doesn't change
    "a3": [5.0, 6.0]
}

# Generate the range of values for each parameter using the stride
for key, val in change_rate_params.items():
    if type(val) is float:
        change_rate_params[key] = [val]
    elif len(val) == 1:
        continue
    elif val[0] == val[1]:
        change_rate_params[key] = [val[0]]
print(change_rate_params)
param_values = {key: np.arange(*values, step=param_stride) if len(values) == 2 else values for key, values in change_rate_params.items()}
# Use itertools.product to generate all combinations of parameter values
combinations = list(itertools.product(*param_values.values()))

# Iterate through each combination
for combination in combinations:
    # Unpack the combination into individual parameters
    strategy_params = {key: value for key, value in zip(change_rate_params.keys(), combination)}
    
    # Now you have the current combination of parameters in strategy_params
    # You can use this dictionary to create a StrategyTurtle instance or any other processing

    # Print the current combination (for demonstration purposes)
    print(strategy_params)
```
